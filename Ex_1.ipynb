```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D', 'E']
values_bar = [23, 45, 56, 78, 34]
values_line = [10, 20, 30, 40, 50]

plt.figure(figsize=(8, 6))
plt.bar(categories, values_bar, color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 6))
plt.plot(categories, values_line, marker='o', color='green', linestyle='-')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Line Chart')
plt.grid(True)
plt.show()
```
